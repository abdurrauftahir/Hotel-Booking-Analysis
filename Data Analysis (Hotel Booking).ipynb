# Data Analysis & Data Visualiation

```python
canceled_percentage=df['is_canceled'].value_counts(normalize=True)# "value_counts()" used for shoowing column name with values counting
canceled_percentage
#canceled_percentage*100
```

```python
#Visualization
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count') #used for giving diagram name
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.8)
plt.show()
```

```python
plt.figure(figsize=(8,4))
axl=sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in different Hotels', size=20) #to giving name/title to plot/diagram
plt.xlabel('hotel') #to giving name of X-axis
plt.ylabel('No of Reservations') # to giving name of Y-axis
plt.legend(['Not Canceled', 'Canceled']) #used for to show Canceled/Not Canceled color scheme to show bars
plt.show()
```

```python
#To check percentage of 'canceled reservation' at Resort Hotel
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#To check percentage of 'canceled reservation' at City Hotel
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#To check "reservation_status" for resort_hotel & city_hotel
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#Average Daily Rate in City and Resort Hotel
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month #to show months
plt.figure(figsize=(16,8))
axl=sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status per Month', size=20) #to giving name/title to plot/diagram
plt.xlabel('Month') #to giving name of X-axis
plt.ylabel('No of Reservations') # to giving name of Y-axis
plt.legend(['Not Canceled', 'Canceled']) #used for to show Canceled/Not Canceled color scheme to show bars
plt.show()
```

```python
#ADR per Month for "Cancelation"
plt.figure(figsize=(12,6))
sns.barplot('month', 'adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title('ADR for each Month', size=18) #to giving name/title to plot/diagram
plt.show()
```

```python
# Cancelation rate on the base of country
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(6,6))
plt.title('Top 10 Countries wit reservation cancelled', size=15)
plt.pie(top_10_country, autopct='%.2f', labels=top_10_country.index)
plt.show()
```

```python
# Offline travel agents to make their reservations
# show "market_segment" to check details of Travel agents(TA)
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
#To check cancelled reservations
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
#To check price/rate of reservation


#For cancelled_data
# grouped for 'resrervation_status_date' andd find Mean of 'ADR'
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True) #reset/remove index
#sort to Date wise "Average Daily Rate (ADR)" in increasing order
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

#For not_cancelled_data
not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)


plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```

```python
#To Filter data year wise like: 2016 to 2017-09
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```
